## Identification and characterization of circular RNAs in Arabidopsis

### Identification pipeline
note: This project began at the end of 2013. There was no available tools then. So Dr.Fu in my lab built this pipeline. I fixed the bugs in Fu's pipeline and added short cRNA identification pipeline. I also modified the pipeline of Memczak(2013) and Zhang(2014) to make them fit the pair-end strand-specific RNase+ data.


```python
display.Image("notebook_material/pipeline-0.jpeg")
```

** This pipeline could be used for Pair-end strand-specific RNase+ data. We could identify cRNA flanking GTAG or not.** 
In following analysis, we found that non-GTAG flanking cRNAs was the major type of cRNA in Arabidopsis. However, There is a problem for non-GTAG flanking cRNA identification: **Shift size**.The "shift problem" of circRNAs boundary were the major obstacle for the identification of non-GTAG circRNAs.(The shift was caused by the beginning of the back splice region was very similar to the end of the back splice region). 1-12 bp shifting was an common phenomenon in the non-GTAG circRNAs.
